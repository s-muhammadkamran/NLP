# Stopwords with Lemmatization

```python
import utility_func as util
import nltk
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer, WordNetLemmatizer
from nltk.tokenize import sent_tokenize
```

```python
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
corpus = util.readfile('../data/stop_words.md')

docs = nltk.sent_tokenize(corpus)

docs
```

```python
wnl = WordNetLemmatizer()
```

```python
stop_words_set = set(stopwords.words('english'))
stop_words_set
```

```python
lemma_sent = []
for i, doc in enumerate(docs):
    words = nltk.word_tokenize(doc.lower())
    words = [wnl.lemmatize(word, pos='v') for word in words if word not in stop_words_set]
    lemma_sent.append(' '.join(words))

lemma_sent
```
